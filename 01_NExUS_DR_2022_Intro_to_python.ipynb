__2) Basic Data Types__

In programming we can work with a multitude of data types, but the main ones are integers, float, string and booleans. Both integers and float are numeric data types, where integers are whole numbers and float are 'floating point' numbers with decimal places. Both integers and float values can be positive, negative or zero. A string is a text data type, like we've seen in the cell above, where the characters are enclosed in either 'single quotations' or "double quotations". A boolean is a binary data type either denoted as True or False. The code cell below has some examples for each of these.

```python
# 1. Integer (int), whole numbers
my_int = 255

# 2. Float, a numeric value with decimal 
my_float = 33.33

# 3. String (str), text value surrounded by either 'these' or "these" 
my_str = "memes"

# 4. Boolean (bool), either True or False
my_bool = True
```

If we are unsure on any of the data types, we can use the built in python function 'type' to check. Below we check what the data type is for the variable ___my_str___. Feel free to substitute the variables out and check them all.

```python
#Printing the datatype of variable 
print(type(my_str))
```

Like most programming languages, python can intuitively work with these different data types and can combine/convert them when neccessary. For example, in the next cell we sum the integer value and the float value together. What we get is a resulting float value.

```python
# Print the sum of the integer and the float value
print(my_int + my_float)
```

__3) Lists__

Another important concept to know in python are lists. Lists are used to store collections of data that can be accessed via a single variable. In the next cell we create a list of rock names and assign it to the variable 'my_rocks'. We can see that the rock names are seperated by commmas and are within single quotation marks, denoting that they are of data type 'string'. Using square brackets and an index number at the end of the list variable name, we can extract the respecitive list value. Note that python uses zero-based indexing, mean that the index 0 is the first item within that list.


```python
# Create a list of rock names
my_rocks = ['Gabbro', 'Basalt', 'Dolerite', 'Andesite', 'Granite', 'Rhyolite']

# Print the 3rd item in the list
print(my_rocks[2])
```

Other important data structures for storage are dictionaries, sets, tuples and arrays. The further you continue into your programming journey, the more familiar with these you will become.

__4) Loops and If-Else Statements__

Loops and If-Else statements are key components in the programming world. Loops provide a way to iteratively access values in a data structure (ie like a list) and If-Else statements provide logicical conditions to your code flow. There are two kinds of loops '_For_' and '_While_' loops. In the cell we set up a '_For_' loop that basically reads "For each rock in my list of rocks, print my rock is a cool rock". __Note__ that because the variable ___rock___ within the '_For_' loop is a string, we can combine it with the string ' is a cool rock!' using a + symbol.

```python
# Loop through rocks list
for rock in my_rocks:

  # Print statement
  print(rock + ' is a cool rock!')
```

We have sucessfully looped through the list of rocks and concluded that they are all 🆒🪨's. In the next cell we make things a little more complicated by adding in an If-Else statement. Here we say that if the rock name is equal to Dolerite, then print this rock is the best rock. If the rock name is not equal to dolerite then do something else. Note the use of double equal signs (==), we use this because a single equal sign (=) is used to set variables. 

```python
#Loop through rocks list
for rock in my_rocks:

  # Set if conditional, and
  # If the rock name in the list is a dolerite run this code
  if rock == 'Dolerite':
    print(rock + ' is the best rock!')
  
  # Otherwise run this code
  else:
    print(rock + ' is a cool rock!')
```

__5) Functions__

We can start to see how introducing more logic can very quickly make our code long and difficult to follow. To aid with this, we can use functions. Functions are a large part of programming and are designed to do specific tasks in a repeatable fashion. In the following cell we have a function that can take geochemical assay values reported in ppb (parts per billion) and convert them to ppm (parts per million). We define a function in python by typing ___def___ followed by the name of the function and then the inputs (arguments) in brackets. In our case both the input and output are numerical values (integer OR float). 

```python
# Defining a function called 'convert_ppb_to_ppm'
def convert_ppb_to_ppm(value_ppb):

  # Converting ppb value to ppm
  value_ppm = value_ppb/1000
  
  return value_ppm
```

When we run the above cell we do not get an output, but the function is now stored in memory and we can call upon it in future. We can test how this works in the cell below. Firstly, we create a new variable called assay_result and then set it as an input it into our function, outputting to a variable called _assay_result_ppm_. We can check the new variable value to see if the function has run correctly.

```python
#Setting an example assay value
assay_result = 66.5

#Convert the value to ppm using our new function
assay_result_ppm = convert_ppb_to_ppm(assay_result)

#Print assay results
print(assay_result_ppm)
```

The function has correctly run and has output a value of 0.0665ppm. This might seem like a bit of a trivial function, but if we wanted to get more complex and began to combine loops, lists and If-Else statements we could extend this function to search through entire datasets and convert all elements with ppb values to ppm all in a single line of code! Functions are extremely useful but are often time consuming and difficult to both write and test. Thankfully there are a plethora of existing functions at our fingertips in the form of python 'libraries'.

There is a global community of programmers who write, test and continually update these libraries to make all our coding lives easier. Libraries generally provide functions for a specific purpose such as visualisation or data manipulation. Some libraries are even large enough to have their own modules within them (like sections within an actual library like science, history, fiction etc) that we can call upon directly. In the cell below we import libraries into out notebook, giving us access to all the functions within them. 

```python
# Importing a library
import numpy

# Importing a library with an alias
import pandas as pd

# Importing a specific module from a library
from sklearn.cluster import KMeans
```

Here we import the libraries numpy, pandas and the Kmeans module from sklearn. Notice that for pandas we import it as pd. This is called an alias and can use the shorthand pd to call the functions.
